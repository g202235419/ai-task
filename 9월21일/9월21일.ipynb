```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel ('C:/Users/mandujoayo/데처프/10월5일/04_data1.xlsx', index_col='area')
data.head()
```

```python
data2020_good = data['2020_good']
data2019_good = data['2019_good']
```

```python
plt.figure(figsize=(20,4))

plt.plot(data2020_good, color='b', marker='o') 
plt.plot(data2019_good, color='r', marker='o')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 Fine Dust (pm10) Good Line Graph')
plt.legend()
plt.grid() 
plt.show()
```

```python
plt.figure(figsize=(20,4))

for year in range (2015, 2021):
    year = str(year)+'_good'
    chartdata = data[year]
    plt.plot(chartdata, marker='o', label=year)
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015-2020 Fine Dust (pm10) Good Line Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib. pyplot as plt

data2015_verybad = data['2015_verybad']
data2020_verybad = data['2020_verybad']

plt.figure(figsize=(20,4))
plt.bar(data2015_verybad. index, data2015_verybad, color= 'g', alpha = 0.5) 
plt.bar(data2020_verybad. index, data2020_verybad, color= 'b') 
plt.xlabel ('Area')
plt.ylabel ('Micrometer')
plt.title('2015 Fine Dust(pm10) Very Bad Bar Graph') 
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange(7)

plt.figure(figsize=(20,4))

data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good'] 
for year in range(2015, 2021):
    year = str(year)+'_good'
    chartdata = data6[year]
    plt.bar(index, chartdata, width=0.15, label=year)
    index = index + 0.15
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Daegu', 'Incheon','Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015-2020 Fine Dust (pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
import numpy as np
index= np.arange(7)

plt.figure(figsize=(10,4))

data6 = data.loc['Seoul': 'Ulsan', '2017_common': '2020_common']
for year in range (2017, 2021):
    year = str(year)+'_common'
    chartdata = data6[year]
    plt.barh(index, chartdata, label=year)

plt.xlim(120,310)
plt.ylabel('Area')
plt.xlabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju',
'Daejeon', 'Ulsan'])
plt.title('2017-2020 Fine Dust (pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/mandujoayo/데처프/10월5일/04_data2.csv')
```

```python
data
```

```python
PetalLength = data['PetalLength'] 
PetalWidth = data['PetalWidth']
plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Petal Scatter Graph')
plt.grid() 
plt.show()
```

```python
PetalLength = data['PetalLength'] 
PetalWidth = data['PetalWidth']

SepalLength = data['SepalLength'] 
SepalWidth = data['SepalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth, color='r', edgecolor='w', label='PetalLength PetalWidth')
plt.scatter (SepalLength, SepalWidth, color='b', edgecolor='w',label='SepalLength * SepalWidth')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Iris Petal Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
versicolor_PetalLength = data.loc[data.Species== 'versicolor', ['Species', 'PetalLength']]
virginica_PetalLength = data.loc[data.Species=='virginica', ['Species', 'PetalLength']]
plt.figure(figsize=(10,6))
plt.hist(versicolor_PetalLength ['PetalLength'], bins=5, alpha=0.5, label='versicolor')
plt.hist(virginica_PetalLength ['PetalLength'], bins=5, alpha=0.5,
label='virginica')
plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Petal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
versicolor_SepalWidth = data.loc[data.Species== 'setosa', ['Species', 'SepalWidth']]
virginica_SepalWidth = data.loc[data.Species=='versicolor', ['Species', 'SepalWidth']]
plt.figure(figsize=(10,6))
plt.hist(versicolor_SepalWidth ['SepalWidth'],color = 'r' ,bins=5, alpha=0.5, label='setosa')
plt.hist(virginica_SepalWidth ['SepalWidth'],color = 'b', bins=5, alpha=0.5, label='versicolor')
plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Petal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```
