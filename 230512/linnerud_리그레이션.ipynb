```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
lin = datasets.load_linnerud()
```

```python
lin
```

```python
df_data = pd.DataFrame(lin.data, columns = lin.feature_names)
df_labels = pd.DataFrame(lin.target, columns = lin.target_names)
```

```python
data = pd.concat([df_data, df_labels.Pulse], axis=1)
```

```python
data
```

```python
X1 = data.iloc[:,0:3].values
y1 = data.iloc[:,3].values

from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X1,y1,test_size=0.2,random_state=0) 
```

```python
X1
```

```python
y1
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(120, activation = 'elu', input_shape = (3,)))
model.add(Dense(100, activation = 'elu'))
model.add(Dense(60, activation = 'elu'))
model.add(Dense(32, activation = 'elu'))
model.add(Dense(1)) 
model.compile(optimizer = 'rmsprop', loss = 'mae', metrics = ['mae'])

model.summary()
```

```python
N_EPOCHS = 100
history = model.fit(X_train, y_train, epochs = N_EPOCHS, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test, y_test)
```
